Los algoritmos de machine learning utilizados en este trabajo son los siguientes:
* **Regresión lineal:** La regresión lineal es un algoritmo de aprendizaje automático supervisado que se utiliza para predecir un valor continuo a partir de una serie de variables independientes.
* **Aprendizaje automático supervisado:** El aprendizaje automático supervisado es un tipo de aprendizaje automático en el que se proporciona al modelo una serie de ejemplos etiquetados. El modelo utiliza estos ejemplos para aprender a clasificar nuevos datos.
* **Aprendizaje automático no supervisado:** El aprendizaje automático no supervisado es un tipo de aprendizaje automático en el que no se proporciona al modelo ninguna información sobre cómo clasificar los datos. El modelo debe aprender por sí mismo a clasificar los datos.
* **K-means clustering:** El clustering K-means es un algoritmo de aprendizaje automático no supervisado que se utiliza para agrupar datos en un número determinado de grupos.
* **SVM:** El SVM (Support Vector Machine) es un algoritmo de aprendizaje automático supervisado que se utiliza para clasificar datos. El SVM funciona separando los datos en dos clases mediante una línea de decisión.

# Visualización.

###¿Como evolucionan los ingresos de la industria cinematográfica y que singularidades podemos encontrar?

```python
df_grouped = IMDb_clean.groupby('Release Year')['Gross worldwide (in millions)'].sum()
plt.figure(figsize=(15, 8))
df_grouped.plot.bar(rot=0)
x = df_grouped.plot.bar(rot=90)
plt.xlabel('Year')
plt.ylabel('Gross worldwide')
plt.title('Gross worldwide (in millions) per year.')
plt.show()
```

Este grafico se puede ver como evolucionan los ingresos año a año a nivel mundial(incluye USA/Canada)

```python
df_grouped2 = IMDb_clean.groupby('Release Year')['Gross in US & Canada (in millions)'].sum()
plt.figure(figsize=(15, 8))
df_grouped2.plot.bar(rot=0)
x = df_grouped2.plot.bar(rot=90)
plt.xlabel('Year')
plt.ylabel('Gross worldwide')
plt.title('Gross in US & Canada (in millions) per year.')
plt.show()
```

Este grafico se puede ver como evolucionan los ingresos año a año en la región Norteamericana ( USA/Canadá).

```python
IMDb_clean['Gross Rest of World (in millions)'] = IMDb_clean['Gross worldwide (in millions)'] - IMDb_clean['Gross in US & Canada (in millions)']

# Lo que hacemos aca es un resta para poder calcular cuanto son los ingresos calculados del resto del mundo (sin USA/Canada). En el dataset se calcula a nivel mundial sumando
# estos dos grupos. Para el analisis posterior es importante tenerlos separados.
# Luego, agrupamos los datos por año y calculamos la suma de los ingresos para cada grupo

grouped = IMDb_clean.groupby('Release Year').sum(numeric_only=True)


plt.figure(figsize=(10, 6))
plt.plot(grouped.index, grouped['Gross in US & Canada (in millions)'], label='US & Canada')
plt.plot(grouped.index, grouped['Gross Rest of World (in millions)'], label='Rest of World')

plt.xlabel('Year')
plt.ylabel('Gross (in millions)')
plt.title('Gross by Year and Region')
plt.legend()
plt.show()
```

Acá podemos ver a principios del siglo actual los ingresos en el resto del mundo comienzan a superar a los generados en la región norteamericana y si mantienen hasta la pandemia de COVID. A la salida de este evento también se nota una diferencia importante favorable al mercado extra-americano.

```python
df_grouped = IMDb_clean.groupby('Release Year')['Title'].count()
plt.figure(figsize=(15, 8))
df_grouped.plot.bar(rot=0)
x = df_grouped.plot.bar(rot=90)
plt.xlabel('Year')
plt.ylabel('Number of movies')
plt.title('Annual premieres')
plt.show()
```

La relación de ingresos entre la región norteamericana y la suma global como con la cantidad de películas lanzadas año a año es palpable. Sería conveniente mirar una matriz de correlación para mostrar con más énfasis esta relación.

```python
df_corr = IMDb_clean.corr(numeric_only=True)
sns.heatmap(df_corr, cmap='Blues',annot=True)
plt.title('Correlation Matrix')
plt.show()
```

El gráfico de correlación muestra que hay una fuerte correlación positiva entre los ingresos brutos mundiales y los ingresos brutos de EE.UU. / Canadá. Esto significa que, si los ingresos brutos mundiales de una película son altos, es probable que sus ingresos brutos de EE.UU. / Canadá también sean altos.
También hay una fuerte correlación positiva entre los ingresos brutos mundiales y el presupuesto de la película. Esto significa que las películas con presupuestos más altos tienden a tener ingresos brutos mundiales más altos.
Dos puntos me llaman la atención:

**Hay una *correlación positiva* entre los ingresos brutos mundiales y la puntuación. Esto significa que las películas con puntuaciones más altas tienden a tener ingresos brutos mundiales más altos.**


**Hay una *correlación negativa* entre los ingresos brutos mundiales y el año de estreno. Esto significa que los ingresos brutos mundiales de las películas tienden a disminuir a medida que pasa el tiempo.**

***Correlacion no sigifica causalidad***, no obstante es interesante graficar estas relaciones.

```python


plt.scatter(IMDb_clean['Rating (Out of 10)'], IMDb_clean['Gross worldwide (in millions)'])
plt.xlabel('Rating (Out of 10)')
plt.ylabel('Gross worldwide (in millions)')
plt.title('Relationship between Rating score and worldwide gross')
plt.show()
```

### Rating e Ingresos.

##### Rating hace referencia a la clasificacion dada por la Motion Picture Association of America (MPAA) en los Estados Unidos.

```python
# Creacion de nueva columna con Rating modificados a int, para poder graficar.

grouped_df = IMDb_clean.groupby('Motion Picture Rating')['Gross in US & Canada (in millions)'].sum().reset_index()

fig = px.bar(grouped_df, x='Motion Picture Rating', y='Gross in US & Canada (in millions)',
             labels={'Motion Picture Rating':'Calificaciones de las películas', 'Gross in US & Canada (in millions)':'Ingreso mundial total (en millones)'}, height=400)
fig.show()
```

Antes analizar el grafico se debe tener en cuenta el sistema de calificación de películas. Este tiene variaciones en los diferentes países. En la industria norteamericana se califican de la siguiente manera:
• G: Audiencias Generales. Todas las edades admitidas.
• PG: Se sugiere la supervisión de los padres. Algunos materiales pueden no ser adecuados para niños pequeños.
• PG-13: Padres fuertemente advertidos. Algunos contenidos pueden ser inapropiados para preadolescentes.
• R: Restringido. Contiene algo de material para adultos.
• NC-17: Claramente Adulto. El contenido es estrictamente para adultos.
Este sistema se creó en 1968, por lo tanto, para películas anteriores a esta fecha poseen otra calificación. Se crea mediante una junta de calificación reunida en la Administración de Clasificaciones y Calificaciones (CARA) que evalúa la película después de que se haya enviado a la Motion Picture Association of America (MPAA). Este sistema no es aplicable al resto del mundo, más allá que sea una “guía” para el resto de la industria. Para poner un ejemplo China tiene un sistema de clasificación de películas diferente al de los Estados Unidos. A diferencia de los EE. UU., que tiene un sistema de clasificación por edades detallado (G, PG, PG-13, R, etc.), China no tiene un sistema de clasificación por edades equivalente1. En lugar de eso, todas las películas que se muestran en los cines de China deben ser adecuadas para todas las edades. La Administración Nacional de Radio y Televisión de China revisa todas las películas antes de su lanzamiento. Las películas que contienen contenido considerado inapropiado pueden ser censuradas o prohibidas1.
Por este motivo, en análisis de la relación Ingresos/Rating de este dataset, solo es viable en el contexto norteamericano y con películas posteriores a 1968.
Teniendo en cuenta esto anterior y mirando el grafico se puede observar que las películas con calificación PG-13 tienen un mayor ingreso acumulado que el resto. Esto obviamente puede deberse a diversos motivos, el principal seria la cantidad de películas emitidas con esa calificación que empujen el número hacia arriba. Lo vamos a ver en el próximo gráfico.




```python
grouped_df = IMDb_clean.groupby('Motion Picture Rating')['Title'].count().reset_index()

fig = px.bar(grouped_df, x='Motion Picture Rating', y='Title',
             labels={'Motion Picture Rating':'Calificaciones de las películas', 'Title':'Cantidad de películas'}, height=400)
fig.show()
```

```python


grouped_df = IMDb_clean.groupby('Motion Picture Rating')['Gross in US & Canada (in millions)'].sum().reset_index()

# Utilizo el groupby armado anteriormente y divido en dos variables las peliculas 'PG-13' y 'R'.

df = grouped_df.loc[grouped_df['Motion Picture Rating'] == 'PG-13']
df2 = grouped_df.loc[grouped_df['Motion Picture Rating'] == 'R']

# Luego hacemos esta ecuación y logramos que porcentaje mayor de ganancias tiene las películas PG - 13 sobre las 'R' (en la región norteamericana).

diff = (df['Gross in US & Canada (in millions)'].sum() - df2['Gross in US & Canada (in millions)'].sum()) / df2['Gross in US & Canada (in millions)'].sum() * 100
print(f'Las películas PG-13 tienen un {diff:.2f}% más de ganancias en EEUU que las calificadas como R.')
```

Como se puede ver, con menos películas lanzadas y mayor ingreso total PG – 13 es la calificación mas rentable (por lo menos en la region norteamericana) . Es un 64% mayor que la categoría que le sigue (que es R). Este análisis no hace referencia a la “calidad” fílmica del producto, sino meramente a la rentabilidad del mismo.

Intentare ahora generar un modelo de clasificación utilizando KNN.
La idea es utilizar las clasificaciones realizadas por las personas mostradas en Rating (Out of 10) pero traduciendo estas a buenas o malas películas. Necesitare generar columnas nuevas utilizando funciones propias de las distintas librerías (como padas o sklearn).


###Modificacion de datos y creacion de columnas.

```python
# En este codigo, creamos una nueva columna llamada "target". Esta columna toma los "Rating (Out of 10)" y crea tres valores, Muy bueno ( valores de 10 a 8), buenos ( valores de 7  a 5)
# y malo ( de 4 a 0)
for index, row in IMDb_clean.iterrows():
  if row['Rating (Out of 10)'] >= 6:
    IMDb_clean.at[index, 'targets'] = 0 # buena pelicula
  else:
    IMDb_clean.at[index, 'targets'] = 1 # mala pelicula


print("Cantidad de peliculas buenas: ", len(IMDb_clean[IMDb_clean['targets'] == 0]))

print("Cantidad de peliculas malas: ", len(IMDb_clean[IMDb_clean['targets'] == 1]))
```

```python
# Algunas de las columnas tienen información en forma de cadenas de texto.
# Para utilizar K Nears necesito que estas columnas sean transformadas a número.
# La función factorize lo hace, a la vez creo una nueva columna con dicha información.
# Esta será luego utilizada como vector en el modelo.

IMDb_clean['Director_number'] = IMDb_clean['Director'].factorize()[0]
IMDb_clean['MainGenres_number'] = IMDb_clean['Main Genres'].factorize()[0]
IMDb_clean['Motion Picture Rating_ number'] = IMDb_clean['Motion Picture Rating'].factorize()[0]
```

###Eleccion de vectores.

```python


X = IMDb_clean[['Director_number', 'MainGenres_number', 'Motion Picture Rating_ number','Gross worldwide (in millions)']]
y = IMDb_clean['targets']

# Se divide en datos de entrenamiento y testo.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)






# Este grafico me permite ver que numero de k seria el que presenta mayor Exactitud.


k_values = list(range(1, 17))
scores = []


for k in k_values:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train.values.ravel())
    y_pred = clf.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    scores.append(score)

plt.figure(figsize=(12, 6))
plt.plot(k_values, scores, marker='o')
plt.title('Exactitud del modelo KNN para diferentes valores de k')
plt.xlabel('k')
plt.ylabel('Exactitud')
plt.show()
```

###Modelo, matriz y analisis.

```python

# Elección de verctores.

X = IMDb_clean[['Director_number', 'MainGenres_number', 'Motion Picture Rating_ number','Gross worldwide (in millions)']]
y = IMDb_clean['targets']

# Se divide en datos de entrenamiento y testo.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Se crea el modelo con el k mas apropiado ( uso un k = 6 segun el grafico anterior)

knn = KNeighborsClassifier(n_neighbors=6)

# Se entrenada con datos de entrenamiento y target de entrenamiento.

knn.fit(X_train, y_train)

# Con el modelo entrenado predecimos con la parte de testo, así obtenemos los y predichos.

y_pred = knn.predict(X_test)

# Por último, vemos los resultados comparativos con los targets predichos y los targets reales para ver si el modelo predice bien entre buenas y malas películas.

c = confusion_matrix(y_test, y_pred)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion matrix:')

print('Classification report:')
print(classification_report(y_test.to_numpy(), y_pred))


plt.figure(figsize=(10, 10))
sns.heatmap(c, annot=True, fmt='g', xticklabels=['0', '1'], yticklabels=['0', '1'])
plt.xlabel('Predicho')
plt.ylabel('Verdadero')
plt.title('Confusion matrix')
plt.show()
```

**Precisión (Precision):** La precisión para la clase "buena película" es del 77%, y para la clase "mala película" es del 45%. Esto indica la proporción de predicciones positivas correctas sobre el total de predicciones positivas.


**Recall (Sensibilidad):** El recall para la clase "buena película" es del 97%, y para la clase "mala película" es del 8%. Esto indica la proporción de instancias positivas correctamente identificadas respecto al total de instancias positivas.


**F1-Score:** El F1-score es una métrica que combina precisión y recuperación en un solo valor. Para la clase "buena película" es del 86%, y para la clase "mala película" es del 14%.

**Promedio Macro:** Es el promedio no ponderado de precision, recall y F1-score para ambas clases. En este caso, es del 61%, lo que indica un rendimiento medio.


**Promedio Ponderado:** Es el promedio ponderado por el soporte (número de instancias) de cada clase. En este caso, es del 70%, indicando un rendimiento general del modelo.


En resumen, el modelo tiene una precisión aceptable para la clase "buena película" pero tiene dificultades para clasificar correctamente la clase "mala película". El bajo recall para la clase "mala película" indica que el modelo tiende a pasar por alto muchas instancias de películas malas. Lo que creo que puede afectar el modelo son un par de cosas:

La cantidad de datos, las distintas variables que no ayudan a identificar malas películas y desbalanceados (ya que del total de testeo tengo 3488 películas calificadas como buenas y 1028 como malas).

En resumen podria decir que el modelo no es muy "fuerte", pero tampoco es terrible, ya que tiene problemas en calificar las malas peliculas como malas, pero no asi con las buenas.


```python
# Se calcula la puntuación de validación cruzada para el modelo.

scores = cross_val_score(knn, X, y, cv=5)

# Se imprime la puntuación media.

print('Puntuación media de validación cruzada:', np.mean(scores))
```

El bajo recall para la clase "mala película" indica que el modelo tiende a pasar por alto muchas instancias de películas malas. Evidentemente lo arriba analizado termina siendo probablemente correcto, los tres puntos siguientes al fin y al cabo terminan siendo esenciales:

* La cantidad de datos. El conjunto de datos es relativamente pequeño, y esto puede estar afectando el rendimiento del modelo.
* Las distintas variables que no ayudan a identificar malas películas. Algunas de las variables que se utilizan para entrenar el modelo no son muy útiles para identificar malas películas. Por ejemplo, la variable "Gross worldwide (in millions)" no es muy útil para identificar malas películas, ya que muchas malas películas también tienen un alto valor de ingreso a nivel mundial.
* El desbalanceo de los datos. El conjunto de datos está desbalanceado, ya que hay muchas más películas buenas que películas malas. Esto puede estar afectando el rendimiento del modelo, ya que es más difícil clasificar correctamente las películas malas que las películas buenas.
